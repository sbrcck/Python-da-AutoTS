AutoTS, otomatik zaman serisi tahmini görevi için geliştirilmiş Python'da otomatik bir makine öğrenimi kitaplığıdır. Bu kitaplığı, sonraki n gün sayısı için hisse senedi fiyatlarını tahmin etmek gibi herhangi bir zaman serisi tahmini görevi için kullanabilirsiniz. Bu yazıda, Python'daki AutoTS kitaplığı hakkında bir eğitimden geçireceğim

### Python'da AutoTS nedir?

AutoTS, Otomatik Zaman Serisi anlamına gelir, Python'da zaman serisi tahmini görevi için kullanılan bir makine öğrenme kitaplığıdır. Geçenlerde Apple'ın önümüzdeki 10 gün için hisse senedi fiyatlarını tahmin etmek için kullandım ve ortaya çıkan rakamlar çok doğruydu. Python'daki AutoTS kitaplığının popüler özelliklerinden bazıları şunlardır:

       1) Kullanmakta olduğunuz veri türüne bağlı olarak en iyi zaman serisi tahmin modelini bulmak için kullanılabilir.
       2) Hem tek değişkenli hem de çok değişkenli zaman serilerini işleyebilir.
       3) Ayrıca NaN değerlerini kaldırarak ve doldurarak dağınık verileri işleyebilir ve ayrıca aykırı değerleri de            işleyebilir.
       4) Bu Python kitaplığı tarafından sağlanan modelleri dağıtım için de kullanabilirsiniz.

Bunlar Python'daki AutoTS kitaplığının önemli özelliklerinden bazılarıydı. Zaman serisi tahmini için daha fazla özelliğe sahiptir, bu bilgilere kendi sitesinden öğrenebilirsiniz

Bu Python kütüphanesini daha önce zaman serisi tahmini görevi için hiç kullanmadıysanız, pip komutunu kullanarak kolayca sisteminize kurabilirsiniz; pip install autots. Şimdi, hisse senedi fiyatı tahmini görevi için Python'da AutoTS kitaplığının nasıl kullanılacağını görelim. Gerekli Python kitaplıklarını ve veri setini içe aktararak bu göreve başlayacağım

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('AAPL.csv')
```

Bu derste, Apple'ın  https://query1.finance.yahoo.com/v7/finance/download/AAPL?period1=1587284575&period2=1618820575&interval=1d&events=history&includeAdjustedClose=true ' den indirilen hisse senedi fiyat verilerini kullanıyorum . Şimdi hisse senedi fiyat tahmini görevi için AutoTs kütüphanesini kullanmadan önce, verileri hazırlayalım ve Apple'ın kapanış hisse senedi fiyatlarına bir göz atalım

```python
data = data[["Date", "Close"]]
data["Date"] = pd.to_datetime(data.Date)
data["Close"].plot(figsize=(12, 8), title="Apple Stock Prices", fontsize=20, label="Close Price")
plt.legend()
plt.grid()
plt.show()
```

Şimdi, otomatik zaman serisi tahmini görevi için AutoTS kitaplığının nasıl kullanılacağını görelim

```python
from autots import AutoTS
model = AutoTS(forecast_length=5, frequency='infer', 
               ensemble='simple', drop_data_older_than_periods=200)
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
```

Yukarıdaki kod bölümünde tahmin_uzunluk (forecast_length) parametresini 5 yani hisse senedi fiyatlarını tahmin etmek istediğimiz dönem olarak belirledim. Yani çıktıda, önümüzdeki 10 gün için Apple'ın kapanış hisse senedi fiyatlarını alacağız. Öyleyse tahmin işlevini kullanalım ve çıktıya bir göz atalım

```python
prediction = model.predict()
forecast = prediction.forecast
print("Stock Price Prediction of Apple")
print(forecast)
```
